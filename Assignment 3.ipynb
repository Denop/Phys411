```python
import numpy as np
import math
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as stats

#---------Question One-------------#
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb = hourdata[1,:]
lags = np.arange(720.)
cdcdc = 0.*lags
dc = dc[~np.isnan(dc)]#removing all of the nans
dc = dc - np.mean(dc)#centering on the mean
print(dc.size)
for i,tau in enumerate(lags):#calculating the lag covariance
    if tau==0:
        cdcdc[i]=np.mean(dc*dc)
    else:
        cdcdc[i]=np.mean(dc[:-tau]*dc[tau:])
pdcdc = cdcdc/np.var(dc)
fig,ax=plt.subplots(1,1)
ax.plot(lags,pdcdc)
ax.set_xlabel(r'$\tau [h]$')
ax.set_ylabel('$P_{dcdc} [V^2]$')
#there is a downward trend, the maximum values for the correlation coefficient drop down from 1
#to roughly 0.65 as tau approaches 720.
```

```python
winter = dc[0:2191:1]
summer = dc[4381:6571:1]
cww = 0.*lags
css = 0.*lags
for i,tau in enumerate(lags):#calculating the lag covariance
    if tau==0:
        cww[i]=np.mean(winter*winter)
        css[i]=np.mean(summer*summer)
    else:
        cww[i]=np.mean(winter[:-tau]*winter[tau:])
        css[i]=np.mean(summer[:-tau]*summer[tau:])
pww = cww/np.var(dc)
pss = css/np.var(dc)
fig,ax=plt.subplots(1,1)
ax.plot(lags,pww,label='Winter')
ax.plot(lags,pss,label='Summer')
ax.legend(loc=0)
ax.set_xlabel(r'$\tau [h]$')
ax.set_ylabel('$P_{xx} [V^2]$')
#The summer lag correlation has a much smaller amplitude than the winter one.
#The winter lag correlation trends downwards but the summer one seems to have
#wobble to it and it is un clear if it will continue to trend downward like the
#winter curve.
#The two curves are in phase.
```
